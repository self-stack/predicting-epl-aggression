```python
import pandas as pd
import numpy as np
import df_build as build
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
team = 'Burnley'
df = build.df_team_focus(build.csv_to_df(), team=team)
```

```python
full_df = build.csv_to_df()
teams = build.csv_to_df().HomeTeam.unique()
```

```python
defaultdict(tuple)
```

```python
rc_count_d = defaultdict(list)
game_count = defaultdict(int)

for t in teams:
    rc_count_d[t] = build.df_team_focus(full_df, team=t).team_reds.sum()
    game_count[t] = len(build.df_team_focus(full_df, team=t))
```

```python
for k,v in game_count.items():
    if v > 200: 
        print(k, rc_count_d[k])
```

```python
df.team_reds[df.home_advan].value_counts()
```

```python
home_rc = df.team_reds[df.home_advan]
away_rc = df.team_reds[~df.home_advan]
home_rc_bool = home_rc.astype(dtype='bool')
away_rc_bool = away_rc.astype(dtype='bool')

plt.figure(num=1, figsize=(10,10), dpi=100)
plt.hist(x=[home_rc[home_rc_bool], away_rc[away_rc_bool]], color=['darkgray','r'])
plt.grid(True)
plt.legend(['Home', 'Away'])
plt.title('{0}\'s Red Card Home vs. Away (\'09-\'19)'.format(team), size=16)
plt.xticks([])
plt.xlim(0.5, 1.5)
plt.ylabel('Total RCs',size=14);
plt.savefig('../img/{0}_rc_dist'.format(team))
```

```python
no_rc, rc = df.team_reds.value_counts()
# rc = team_rc[team_rc == 1]
plt.figure(num=1, figsize=(10,10), dpi=100)

plt.pie(x=[no_rc, rc], colors=['darkgray','r'], explode=[0.1, 0], labels=['No Red Card', 'Red Card'], autopct='%1.1f%%')
plt.title('{0}\'s Red Card Pie Chart (\'09-\'19)'.format(team), size=16)
plt.savefig('../img/{0}_rc_pie'.format(team));
```
